## مثال على وظائف وحدة WaPOR
    
</div>

<div dir="rtl">
توضح خلايا الكود المثال أدناه كيف تم استخدام وحدة WaPOR لتنزيل بيانات WaPOR بالجملة: dekadal ، شهريًا ، وسنويًا ، مرجع البخرنتح من الامتداد المقرر. يتم تحديد المدى بواسطة الصندوق المحيط لملف أشكال حوض الأواش. سيتم جمع البيانات النقطية ومعالجتها مسبقًا إلى الوحدة الصحيحة ملم / فترة (ملم / ديكاد ، ملم / شهر ، أو ملم / سنة على التوالي) ، وحفظها في مجلد الإخراج (data)    
</div>

[مجلد الإخراج](data)


<div dir="rtl">

### تعيين نطاق التنزيل ومجلد الإخراج  
    
</div>

<div dir="rtl">
قم بتغيير قيم xmin و ymin و xmax و ymax إلى حدود التنزيل. إذا لزم الأمر، قم بتغيير مسار output_dir إلى المجلد الذي تريد حفظ البيانات فيه.    
</div>
<div dir="rtl"><br/>
** ملاحظة حول التركيب **: بالنسبة للنصوص الموجودة في هذه المفكرات ، تحتوي نهايات معظم ملفات الإدخال والإخراج على "_fh". fh تعني "file handle" وتشير إلى ملف واحد. قد ترى أيضًا "_fhs" ، والتي تعني "file handle series". يشير هذا إلى سلسلة أو مجموعة من الملفات.<div/>

```python
# حدد مدى التنزيل من ملف الشكل:
shape_fh=r".\data\Awash_shapefile.shp"
shape=shapefile.Reader(shape_fh)
xmin,ymin,xmax,ymax=shape.bbox

# تحديد المدى للتنزيل حسب الإحداثيات:
# xmin,ymin,xmax,ymax=(32.65692516077674,-25.16412729789142,32.90420244933813,-25.01450956754679)

output_dir=r'.\data' #مجلد الإخراج
```

<div dir="rtl">

### تحميل البيانات اليومية
    
</div>

<div dir="rtl">
لاحظ أن البيانات اليومية متاحة فقط للمستوى 1 للبخرنتح المرجعي والهطول    
</div>

```python
WaPOR.download_daily(output_dir, 
                       data='RET',
                       Startdate='2009-01-01', 
                       Enddate='2009-01-31',
                       latlim=[ymin-0.5, ymax+0.5], 
                       lonlim=[xmin-0.5, xmax+0.5],
                       level=1, 
                       )
```

<div dir="rtl">

### تحميل بيانات dekadal (10 أيام)
    
</div>

<div dir="rtl">    
ملاحظة: سيتم تحويل وحدة مجموعة بيانات dekadal (10 أيام) إلى كمية / dekad بدلاً من متوسط الكمية اليومية / اليوم
</div>

```python
WaPOR.download_dekadal(output_dir, 
                       data='RET',
                       Startdate='2009-01-01', 
                       Enddate='2009-12-31',
                       latlim=[ymin-0.5, ymax+0.5], 
                       lonlim=[xmin-0.5, xmax+0.5],
                       level=1, 
                       )
```
